### Data Cleaning 

- Year Data have so, much noise , we need to clean it carefully 
- Car['price'] = have a data name called ask for price 
- kms_driven = have nan values and split it 
- fuel_type = lpg, diesel , petrol and nan

- Price - chnage to int and remove ask for prices

```python
## Car Year cleaning 
## str operation karna hai [.str]
car['year'].str.isnumeric() ## return True for all numberic values 

car = car[car['year'].str.isnumeric()] ## return all filter numberic values 

```

```python
car['year'] = car['year'].astype(int)

## Year work done  
```

```python
car.info()
```

```python
## Price Cleaning 

car=car[car['Price'] != 'Ask For Price']
```

```python
car['Price'] = car['Price'].str.replace(',','').astype(int)
```

```python
car.info();
```

```python
car['kms_driven'].unique()
```

```python
# car['kms_driven'].str.isnumeric()
car['kms_driven']=car['kms_driven'].str.replace(',','').str.split().str[0]
```

```python
car = car[car['kms_driven'].str.isnumeric()]
```

```python
car['kms_driven'] = car['kms_driven'].astype(int)
```

```python
car.info()
```

```python
car = car[~car['fuel_type'].isna()] 
## ~ it do remove this row 

# car['fuel_type'].isna()  this return in boolen 
# car[~car['fuel_type'].isna()]  -- but this return from the data 
```

```python
car.info()
```

```python
car['name'] 
## will take only first 3 words not the whole model 
```

```python
# car['name'].str.split(' ')
car['name'] = car['name'].str.split(' ').str.slice(0,3).str.join(' ')
```

```python
## Index sequence problem beacse we remove values 

car.reset_index(drop=True)
```

```python
# this data have outliers beacuse 75% are under 5 lakh and 25% is near 85 lakh 

# we will take only cars less then 60 lakh 

car = car[car['Price']<6e6].reset_index(drop=True);
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import make_column_transformer
from sklearn.pipeline import make_pipeline
```

```python
ohe = OneHotEncoder()
ohe.fit(x[['name','company','fuel_type']])
ohe.categories_
```

```python
column_trans = make_column_transformer((OneHotEncoder(categories=ohe.categories_),['name','company', 'fuel_type']), remainder='passthrough');
```

```python
lr = LinearRegression();
```

```python
pipe = make_pipeline(column_trans, lr);
```

```python
pipe.fit(x_train,y_train);
```

```python
column_trans
```

```python
y_pred = pipe.predict(x_test);
```

```python
r2 = r2_score(y_test,y_pred);
```

```python
r2
```

```python
score = []
for i in range(1000):
  x_train, x_test,y_train, y_test = train_test_split(x,y,test_size=0.2,random_state=i)
  lr= LinearRegression()
  pipe = make_pipeline(column_trans,lr)
  pipe.fit(x_train,y_train)
  y_pred = pipe.predict(x_test);
  score.append(r2_score(y_test,y_pred))
```

```python
np.argmax(score)
```

```python
score[np.argmax(score)]
```

```python
x_train, x_test,y_train, y_test = train_test_split(x,y,test_size=0.2,random_state=np.argmax(score))
lr= LinearRegression()
pipe = make_pipeline(column_trans,lr)
pipe.fit(x_train,y_train)
y_pred = pipe.predict(x_test);
r2_score(y_test,y_pred);
```
